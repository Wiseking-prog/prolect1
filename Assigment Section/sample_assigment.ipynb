## 1. Student Grade Management System

### Goal

Build a command-line application that allows teachers to manage student records.

### Features

* Add student
* Remove student
* Update student scores
* Calculate average score
* Find highest and lowest scoring student
* Search by student name
* Save records to a file
* Load records when the program starts

### Skills

* Dictionaries
* Lists
* Functions
* Loops
* File handling
* Error handling
