```python
def exam(question:dict):
    for key,value in question.items():
        print(value.get("question"))

        answer = value.get("answer")
        
        options = value.get("options")

        for key,value in options.items():
            print(f"{key}. {value}")

        user_option = input("Select an option: ")
        

        user_option = user_option.lower
        answer = answer.lower

        if user_option == answer:
            print("wow you got the answer correctly next question")
        else:
            print(f"you failed the the answer is {answer}")

exam(question_bank)
        
```
