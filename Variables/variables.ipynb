#### What is a Variable:
A Variable is Labeled container that holds data in a computer

#### Why do we have Variables in Programming
- For holding values that are unknown when a code is written, so some values changes as the code runs like in a game (player_score) will always change

- Variables are used to make code resuable say we define a variable
(total_price = price + (price*tax_rate)) we can now use the variable total_price any where in our code

#### When do you use Variables

- When Data changes over time: Example when tracking a user's shopping cart total.
- When data is unknown ahead of time: if your code relies on information that does not exist until someone actually uses the program you must create a variable it would act as a placeholder.
- When the value will be reused multiple times.
