Loops are python statements that help us automate repititive tasks

A for loop is a programming construct used to execute a block of code repeatedly. It is highly efficient for running tasks a specific number of times or stepping through collections of data (like lists or arrays) one by one

```python
coins = ["btc", "eth", "sol"]

for coin in coins:
    print(coin)
```
