#### What is a Tuple?

A tuple is a collection of multiple values stored in a single variable.

Like lists, tuples can store:

Numbers
Strings
Different data types
Other collections

The main difference:

Lists can be changed. Tuples cannot be changed.

````python
student = ("Joseph", 20, "Computer Science")
#    ↓                  ↓
#variable name        Tuple
````

##### Why do we have tuples in programming?
Sometimes we have data that should not change after it is created.

Example: when dealing with user coordinates
`location = (6.5244, 3.3792)`

The latitude and longitiude should stay toghter and should not accidentally be modified

- a tuple protects data




##### When Do You Use Tuples?

1. Storing constant data
2. Returning Multiple Values from a function
3. Stroring coordinates
4. Representing Fixed Record

##### Short Note on Tuple Operations

##### 1. Creating a Tuple

```python
colors = ("red", "blue", "green")
```

Single item tuple:

```python
number = (5,)
```

Without the comma:

```python
number = (5)
```

Python treats it as an integer.

---

##### 2. Accessing Tuple Items

Tuples use indexing.

```python
names = ("John", "Mary", "Alex")

print(names[0])
```

Output:

```text
John
```

---

##### 3. Slicing Tuples

```python
numbers = (1,2,3,4,5)

print(numbers[1:4])
```

Output:

```text
(2,3,4)
```

---

##### 4. Tuple Unpacking

Extract values from a tuple.

```python
person = ("Joseph", 20, "Python")

name, age, skill = person

print(name)
```

Output:

```text
Joseph
```

---

##### 5. Joining Tuples

```python
a = (1,2,3)
b = (4,5,6)

c = a + b

print(c)
```

Output:

```text
(1,2,3,4,5,6)
```

---

##### 6. Repeating Tuples

```python
numbers = (1,2)

print(numbers * 3)
```

Output:

```text
(1,2,1,2,1,2)
```

---

##### 7. Counting Values

```python
numbers = (1,2,2,3)

print(numbers.count(2))
```

Output:

```text
2
```

---

##### 8. Finding Index

```python
numbers = (10,20,30)

print(numbers.index(20))
```

Output:

```text
1
```

---

##### 9. Checking Membership

```python
skills = ("Python", "AI", "SQL")

print("AI" in skills)
```

Output:

```text
True
```
