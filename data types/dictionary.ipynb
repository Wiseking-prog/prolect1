#### What is a Dictionary in python

A Python dictionary is a built-in data structure that stores data in key-value pairs.


It behaves like a real-world dictionary where you look up a word (the key) to find its definition (the value)

#### Why do we have dictionaries in python

- They alllow us to store and look up data using meaningful labels instead of numeric positions like list and tuples

- it allows you to look up information using a specific label rather than a numeric index. 

```python
# how to write dictionaries in python


user_dict  = {"name": "Alice", "age": 30, "city": "Austin"}
#    ↓                           ↓                    ↓ 
#variable name                  key                 value



print(user_dict["age"])
```

Dictionaries also have dictionary opreations that allow you to create, manipulate, and extract data from dictionaries

## 🕵️‍♂️ Mission: The Rogue AI Log Analyzer## Background Story for Students
A rogue AI has scrambled the server logs of a high-security facility. The AI left behind a messy data stream of user actions, but the security team needs a clean report showing exactly who accessed which servers, how many times, and which malicious commands were run.
Your job is to build the central parser using Python dictionaries.
------------------------------
## The Raw Data (Provide this to students)

# Raw, unordered logs from the server roomscrambled_logs = [
    {"user": "Alice", "server": "Alpha", "action": "login", "status": "success"},
    {"user": "Bob", "server": "Beta", "action": "read_file", "file": "config.txt"},
    {"user": "Alice", "server": "Alpha", "action": "run_command", "cmd": "sudo rm -rf"},
    {"user": "Charlie", "server": "Gamma", "action": "login", "status": "fail"},
    {"user": "Bob", "server": "Alpha", "action": "login", "status": "success"},
    {"user": "Alice", "server": "Beta", "action": "login", "status": "success"},
    {"user": "Alice", "server": "Alpha", "action": "run_command", "cmd": "download_data"},
    {"user": "Bob", "server": "Beta", "action": "run_command", "cmd": "sudo rm -rf"},
    {"user": "Charlie", "server": "Gamma", "action": "login", "status": "success"},
]

------------------------------
## Assignment Tasks## Task 1: The Master Security Profile (Nested Dictionaries)
Process scrambled_logs into a single nested dictionary named security_profiles.

* The outer keys must be the usernames.
* The values must be another dictionary tracking:
1. servers_visited: A set (or list) of unique servers they accessed.
   2. command_count: The total number of times they executed a "run_command" action.
   3. red_flags: A list of specific commands run if they contained the phrase "sudo".

## Task 2: Safe Investigation (No Crashes allowed!)
Write a function get_user_status(profiles, username) that takes your security_profiles dictionary and a string name.

* It must return: "{username} ran {count} commands."
* If the user does not exist in the dictionary, it must not crash. Instead, use dictionary methods to safely return: "User {username} not found in system logs."

## Task 3: The Threat Map (Reverse Mapping)
The security team needs a quick way to look up threats by command, rather than by user. Create a new dictionary called threat_map.

* The keys must be the flagged commands (any command containing "sudo").
* The values must be a list of users who ran that specific command.

------------------------------

## Why this is a "Brain Cracker" for Students:

* Initialization Traps: They cannot just assign dict[user]['command_count'] += 1. The user dictionary doesn't exist yet! They must learn to use .setdefault(), if user not in dict, or defaultdict.
* Data Type Mixing: It forces them to store a set (unique servers), an int (counter), and a list (red flags) all inside the same value dictionary.
* Inversion Logic: Task 3 forces them to loop through their newly created dictionary, extract inner lists, and flip the keys and values upside down.
