#### What is a String:

 A string is an immutable sequence of Unicode characters used to represent textual data. 

#### Why do we have Strings in Programming

- We have strings in programming primarily to bridge the gap between how humans communicate (using text) and 

- how computers process data (using numbers)

#### When do you use Strings

- Strings are used for input  processing
-  they are used in text analysis and Natural Language Processing(NLP)
- They are used for Data Validation

#### String Opreattions
##### Python String Operations Cheat Sheet

##### 1. Accessing Characters

```python
word = "Python"

print(word[0])   # P
print(word[1])   # y
print(word[-1])  # n
```

---

##### 2. Slicing Strings

```python
word = "Python"

print(word[0:3])  # Pyt
print(word[2:])   # thon
print(word[:4])   # Pyth
print(word[::-1]) # nohtyP
```

---

##### 3. String Concatenation

Joining strings together.

```python
first = "Hello"
last = "World"

print(first + " " + last)
# Hello World
```

---

##### 4. String Repetition

```python
print("Hi " * 3)

# Hi Hi Hi
```

---

##### 5. Membership Operators

Check if a substring exists.

```python
text = "Machine Learning"

print("Machine" in text)      # True
print("Python" in text)       # False
print("Python" not in text)   # True
```

---

##### 6. String Length

```python
name = "TensorFlow"

print(len(name))
# 10
```

---

##### 7. Changing Case

```python
text = "python programming"

print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
```

Output:

```text
PYTHON PROGRAMMING
python programming
Python Programming
Python programming
```

---

##### 8. Removing Spaces

```python
text = "   Hello World   "

print(text.strip())
print(text.lstrip())
print(text.rstrip())
```

---

##### 9. Replacing Text

```python
text = "I love Java"

new_text = text.replace("Java", "Python")

print(new_text)
# I love Python
```

---

##### 10. Splitting Strings

Convert a string into a list.

```python
sentence = "AI will change the world"

words = sentence.split()

print(words)
```

Output:

```python
['AI', 'will', 'change', 'the', 'world']
```

---

##### 11. Joining Strings

Convert a list into a string.

```python
words = ["AI", "is", "awesome"]

sentence = " ".join(words)

print(sentence)
```

Output:

```text
AI is awesome
```

---

##### 12. Finding a Substring

```python
text = "Python Programming"

print(text.find("Program"))
print(text.index("Program"))
```

Both return:

```text
7
```

Difference:

* `find()` returns `-1` if not found.
* `index()` raises an error if not found.

---

##### 13. Counting Occurrences

```python
text = "banana"

print(text.count("a"))
# 3
```

---

##### 14. Checking String Properties

```python
print("123".isdigit())     # True
print("hello".isalpha())   # True
print("hello123".isalnum())# True
print("HELLO".isupper())   # True
print("hello".islower())   # True
```

---

##### 15. String Formatting

###### f-strings (Recommended)

```python
name = "Joseph"
age = 20

print(f"My name is {name} and I am {age} years old.")
```

Output:

```text
My name is Joseph and I am 20 years old.
```
