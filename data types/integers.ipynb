#### What is an Integer?

This is a Datatype that stores a whole number without any decimal points.

Integers can be 

positive - 1, 
negative -1,
or zero


#### Why does Integers exists

- Computers need a way to store exact values for numbers

#### When do you use Integers

- for controling Loops 
```
for i in range(5):  # Runs exactly 5 times
    print("Hello")

```
- for recording variables with numbers
