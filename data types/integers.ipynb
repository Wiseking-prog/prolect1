Integers can be 

positive - 1, 
negative -1,
or zero


#### Why does Integers exists

- Computers need a way to store exact values for numbers

#### When do you use Integers

- for controling Loops 
```

    print("Hello")

```
- for recording variables with numbers

Write code to:

Collect 3 ticket numbers.

Check that the number is between 1000 and 9999.

If valid, add it to a list of tickets.

if not valid don't add to the list of tickets

Print the list of tickets

````python
numbers = []

while len (numbers) < 3:
    number = int(input("collect 3 ticket nmbers"))

    if number >= 1000 or number <= 9999:
        print("collect it to the list")

        numbers.append(number)

    else:
        print("do not collect the list of tickets")
        

print(numbers)
````
