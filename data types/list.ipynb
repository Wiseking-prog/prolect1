#### What is a List:

A list is a data structure in Python used to store multiple values in a single variable.

A list can store:

- Numbers (Integers)
- Strings
- other lists
- and other datatypes toghter

#### Why do we have Lists in Programming

- whitout list we would need many variables to store related data.

an example will be:

##### Without a Lists:
```
student1 = "John"
student2 = "Mary"
student3 = "David"
```

##### But with Lists:
`students = ["John", "Mary", "David"]`

- They organize related data
- Store collections of items
- Loop through multiple values
- Modify data dynamically




#### When Do You Use Lists?

- for storing user data
- managing products in an online store
- for processing text

##### Short Note on List Operations

##### 1. Creating a List

```python
numbers = [1, 2, 3, 4]
```

---

##### 2. Accessing Items

Lists use indexes.

```python
names = ["John", "Mary", "Alex"]

print(names[0])
# John

print(names[-1])
# Alex
```

---

##### 3. Adding Items

##### append()

Adds an item to the end.

```python
names.append("David")
```

##### insert()

Adds an item at a specific position.

```python
names.insert(1, "Peter")
```

---

##### 4. Removing Items

##### remove()

```python
names.remove("John")
```

###### pop()

Removes using index.

```python
names.pop(0)
```

---

##### 5. Changing Items

```python
names[0] = "Michael"
```

---

##### 6. Finding Length

```python
len(names)
```

---

##### 7. Checking Items

```python
"John" in names
```

Returns:

```python
True
```

---

##### 8. Sorting Lists

```python
numbers = [5, 2, 8, 1]

numbers.sort()

print(numbers)
```

Output:

```python
[1, 2, 5, 8]
```

---

##### 9. Looping Through Lists

```python
for name in names:
    print(name)
```

---

##### 10. List Slicing

```python
numbers = [1,2,3,4,5]

print(numbers[1:4])
```

Output:

```python
[2,3,4]
```
