#### What is a Set?

A set is a collection of unique values in Python.

Unlike lists:

- Sets do not allow duplicates
- Sets are unordered
- Sets are changeable
- Sets use { } brackets

#### Why Do We Have Sets in Programming?

Sometimes we don't care about the order of data. We only care about whether something exists or whether values are unique.

So whitout set's we would have:

```
emails = [
    "a@gmail.com",
    "b@gmail.com",
    "a@gmail.com",
    "c@gmail.com"
]
```

with set's 

```
unique_emails = set(emails)

print(unique_emails)
```

````python
# Here is how to define a set in python

blocked_users = {"john", "mike", "alex"}
#    ↓                     ↓
#variable name            Set
````


#### When Do You Use Sets?

1. For removing duplicate data
2. Fast Membership checking
3. Comparing groups of data

##### Short Note on Set Operations

##### 1. Creating a Set

```python
colors = {"red", "blue", "green"}
```

Empty set:

```python
empty = set()
```

Note:

```python
{}
```

creates a dictionary, not a set.

---

##### 2. Adding Items

##### add()

```python
colors.add("yellow")
```

---

##### 3. Removing Items

##### remove()

```python
colors.remove("red")
```

Raises an error if item doesn't exist.

##### discard()

```python
colors.discard("black")
```

Does not raise an error.

---

##### 4. Union (Combine Sets)

Combines all unique values.

```python
a = {1,2,3}
b = {3,4,5}

print(a | b)
```

Output:

```python
{1,2,3,4,5}
```

---

##### 5. Intersection (Common Values)

Find values that exist in both sets.

```python
print(a & b)
```

Output:

```python
{3}
```

---

##### 6. Difference

Find values that exist in one set but not another.

```python
print(a - b)
```

Output:

```python
{1,2}
```

---

##### 7. Symmetric Difference

Values that exist in either set but not both.

```python
print(a ^ b)
```

Output:

```python
{1,2,4,5}
```

---

##### 8. Checking Membership

```python
"Python" in skills
```

Returns:

```python
True
```

---

##### 9. Set Size

```python
len(colors)
```
