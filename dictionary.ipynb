```python
trade = {"price": 5000,
         "volume": 10,
         "cion": "bitcoin"}
print(trade.keys())

print(trade.values())

print(trade.items())

print(trade.get("price"))

trade.update({"price": 7000})
print(trade)

trade.pop("volume")
print(trade)
```
